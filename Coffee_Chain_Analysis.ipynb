Analysing Sales

```python
print("'Sales'")                                                                
print("Minimum value: ", data["Sales"].min())
print("Maximum value: ", data["Sales"].max())
print("How many values are NaN?: ", pd.isnull(data['Sales']).sum())
```

Calculating the average Profit in each state

```python
Prof = data.groupby('State')['Profit'].mean()
Prof
```

Arranging the states based on their Profit

```python
data.sort_values(by ='Profit',ascending = False).groupby('State')['Profit'].mean()
```

Visualizing sales 

```python
x = data['State'].unique()
y = data.groupby('State')['Profit'].mean()
plt.scatter(x,y,label = 'Sales in different states',color = 'm',s = 40,alpha = 0.8)
plt.legend()
plt.xlabel('State')
plt.ylabel('Profit')
plt.show()
```

```python
x = data['State'].unique()
y = data.sort_values(by = 'Profit',ascending = False).groupby('State')['Profit'].mean()
plt.barh(x,y,label = 'Sales in different states',color = 'm')
plt.legend()
plt.xlabel('State')
plt.ylabel('Profit')
plt.show()
```

```python
#To know which product performs well in profit
data.groupby('Product')['Profit'].mean()
```

```python
x = data['Product'].unique()
y = data.groupby('Product')['Profit'].mean()
plt.barh(x,y,label = 'performance of products',color = 'm')
plt.legend()
plt.xlabel('Product')
plt.ylabel('Profit')
plt.show()
```
